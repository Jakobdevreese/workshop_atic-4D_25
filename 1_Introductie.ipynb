### 1. Virtuele omgeving opzetten

Een **virtuele omgeving** is een geÃ¯soleerde Python-installatie voor je project. Het voorkomt conflicten tussen pakketversies van verschillende projecten en houdt je systeembrede Python-installatie schoon.

**Waarom een virtuele omgeving?**
- Elk project krijgt z'n eigen set pakketten
- Reproduceerbaarheid: andere gebruikers kunnen exact dezelfde omgeving opzetten
- Geen gedoe met versieconflicten tussen projecten

**Opzetten in de terminal:**

```bash

.venv\Scripts\activate                 # Windows (Command Prompt)
.venv\Scripts\Activate.ps1             # Windows (PowerShell)
```
